```python
import pandas as pd
import plotly.express as px

df_1996_2024 = pd.read_csv('data_small_100.csv')
df_1996_2024
```

```python
import pandas as pd
import plotly.express as px

df = df_1996_2024.copy()
fig = px.density_map(df, lat='Lat', lon='Long', radius=10,
                        center=dict(lat=df.Lat.mean(), lon=df.Long.mean()), zoom=10,
                        map_style="open-street-map")
fig.update_layout(width=1200, height=800)
fig.show()
```

[Plotly : scatter_map](https://plotly.com/python-api-reference/generated/plotly.express.scatter_map.html)

```python
df = df_1996_2024.copy().sample(n=50)
df2 = df_1996_2024.copy().sample(n=50)
fig_base = px.scatter_map(df, lat='Lat', lon='Long',
                        center=dict(lat=df.Lat.mean(), lon=df.Long.mean()), zoom=10,
                        map_style="open-street-map",hover_name="Species", hover_data=['Depth','Distance','Group-Size'],labels={'Lat':'Latitude'})

fig_base.update_traces(marker=dict(size=6, color="red", symbol="circle"))

fig_added = px.scatter_map(df2, lat='Lat', lon='Long',
                        center=dict(lat=df.Lat.mean(), lon=df.Long.mean()), zoom=10,
                        map_style="open-street-map",hover_name="Species", hover_data=['Depth','Distance','Group-Size'],labels={'Lat':'Latitude'})

fig_added.update_traces(marker=dict(size=6, color="blue", symbol="circle"))

# Add scatter traces to fig_base
for trace in fig_added.data:
    fig_base.add_trace(trace)

fig_base.update_layout(width=1200, height=800)
fig_base.show()
```
