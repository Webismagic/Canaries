```python
import pandas as pd
import plotly.express as px

df_outliers = pd.read_excel('outliers.xlsx')
df_outliers
```

[Plotly : scatter_map](https://plotly.com/python-api-reference/generated/plotly.express.scatter_map.html)

```python

fig_base = px.scatter_map(df_outliers, lat='Lat', lon='Long',
                        center=dict(lat=df_outliers.Lat.mean(), lon=df_outliers.Long.mean()), zoom=9,color='Species',
                        map_style="open-street-map",hover_name="Species", hover_data=['Depth','Distance','Group-Size','variable','value'],labels={'Lat':'Latitude'})

fig_base.update_traces(marker=dict(size=4, symbol="circle"))

fig_base.update_layout(width=1200, height=800)
fig_base.show()
```
